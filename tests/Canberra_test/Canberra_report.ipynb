```python
import wiki_compressor_diff_match as wc
```

```python
w = wc.wikiConverter()
n = w.get_n('Canberra.xml')
```

```python
n
```

```python
import math
import xml.etree.ElementTree as ET
import time
```

```python
k=[]
i1= int(math.sqrt(n))

if i1<=1000:
    for i in range(i1,1000,100):
        k.append(i)
    k.append(1000)
else:
    k.append(1000)
    for i in range(1000,i1,100):
        k.append(i)

```

```python
k
```

```python
compress_time =[]
decompress_time = []
filename='Canberra.xml'
```

```python
for i in range(len(k)):
    start = time.time()
    w.wikiConvert(file_name=filename,output_dir=filename+'compressed'+str(k[i]),k=k[i])
    end = time.time()
    compress_time.append(end-start)
```

```python
compress_time
```

```python
for i in range(len(k)):
     w.wiki_decompress(filename+'compressed'+str(k[i])+'/Canberra.knolml',k[i],filename+'wiki_decompress'+str(k[i]),decompress_time)
    
```

```python
k,n,compress_time,decompress_time
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 
br1 = np.arange(len(k)) 
br2 = [x + barWidth for x in br1]  

plt.bar(br2, decompress_time, color ='g', width = barWidth, 
        edgecolor ='grey', label ='decompress')
plt.xlabel('k', fontweight ='bold') 
plt.ylabel('time in seconds', fontweight ='bold')
plt.legend(('compress','decompress'),loc='upper center', shadow=True)
plt.xticks([r + barWidth for r in range(len(compress_time))],k) 
```
